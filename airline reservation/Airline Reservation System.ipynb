```python
def checkAvailableSeats(noOfSeats,Class):
    availableSeats=0
    if Class == "BusinessClass":
        for i in seats[0:2]:
            availableSeats= availableSeats+ i.count(0)
                
    elif Class == "FirstClass":
        for i in seats[2:4]:
            availableSeats= availableSeats+ i.count(0)
             
    elif Class == "EconomyClass":
        for i in seats[4:]:
            availableSeats= availableSeats+ i.count(0)
             
    else:
        print("invalid class")
    
    if (availableSeats >= noOfSeats):
        return True
    else:
        return False
    
```

```python
## sum(sum(np.array(eclass) * 10000))

Eclass = 10000
Fclass = 14000
Bclass = 18000

def getFair(Class):
    total = 0
    if Class == "EconomyClass":
        for i in seats[4:]:
            for j in i:
                total = total + (j* Eclass) 
    
    elif Class == "FirstClass":
        for i in seats[2:4]:
            for j in i:
                total = total + (j* Fclass) 

    
    elif Class == "BusinessClass":
        for i in seats[0:2]:
            for j in i:
                total = total + (j* Bclass) 

    print(total)
    
def getOvrFair():       ##for overall fair/ fair for all classes
    total = 0
    
    for i in seats[4:]:
        for j in i:
            total = total + (j* Eclass) 
    
    for i in seats[2:4]:
        for j in i:
            total = total + (j* Fclass) 

    
    for i in seats[0:2]:
        for j in i:
            total = total + (j* Bclass) 

    print(total) 
```

```python

def bookBusinessClass(noOfSeats):
    
    flag = False
    flag = checkAvailableSeats(noOfSeats,"BusinessClass")     # flag tells us that is/are the seats available or not
    
    if flag:
        k=0          ##for adding number of seats
        for i in seats[:2]:
            for j in range(7):
                k+=1
                if k<= noOfSeats:
                    i[j] = 1
        print("Congratulations! You have got your booking in Business class")
        print("Fair of your Booking is : ")
        getFair("BusinessClass")
    if not flag:
        print("Sorry we don't available seats")

    
def bookFirstClass(noOfSeats):
    
    flag = False
    flag = checkAvailableSeats(noOfSeats,"FirstClass")     # flag tells us that is/are the seats available or not
    
    if flag:
        k=0          ##for adding number of seats
        for i in seats[2:4]:
            for j in range(7):
                k+=1
                if k<= noOfSeats:
                    i[j] = 1
        print("Congratulations! You have got your booking in First class")
        print("Fair of your Booking is : ")
        getFair("FirstClass")
    if not flag:
        print("Sorry we don't available seats")

    
def bookEconomyClass(noOfSeats):
    
    flag = False
    flag = checkAvailableSeats(noOfSeats,"EconomyClass")     # flag tells us that is/are the seats available or not
    
    if flag:
        k=0          ##for adding number of seats
        for i in seats[4:]:
            for j in range(7):
                k+=1
                if k<= noOfSeats:
                    i[j] = 1
        print("Congratulations! You have got your booking in Economy class")   
        print("Fair of your Booking is : ")
        getFair("EconomyClass")

    if not flag:
        print("Sorry we don't available seats")
```
